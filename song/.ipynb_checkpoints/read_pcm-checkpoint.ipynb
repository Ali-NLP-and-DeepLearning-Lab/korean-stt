```python
import glob
from tqdm import tqdm_notebook as tqdm
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd
import re
import jamotools
```

```python
data_path = 'D:/aihub/'
```

```python
kspon_speech_paths = glob.glob(data_path + '*')
```

```python
sub_path = []

for path in kspon_speech_paths:
    sub_path += glob.glob(path + '/*')
    
pcm_list = []
txt_list = []

for path in tqdm(sub_path):
    pcm_list += glob.glob(path + '/*.pcm')
    txt_list += glob.glob(path + '/*.txt')
```

```python
def find_parenthesize_index(sentence):
    l_count = sentence.count('(')
    r_count = sentence.count(')')
    
    if l_count == r_count:
        count = l_count
        
        search_result = list()
        
        if count == 0:
            return search_result
        
        search_idx = 0
                
        for i in range(count):
            search_string = sentence[search_idx:]
            try:
                l_idx = search_string.index('(') + search_idx
                r_idx = search_string.index(')') + search_idx
                search_result.append((l_idx, r_idx))
                search_idx = r_idx + 1
            except:
                return search_result
        
        return search_result
            
    else:
        print('Errornous parenthesis found in {}'.format(sentence))
        return None 
```

```python
def remove_parenthesized_word(sentence, mode='odd'):
    idxs = find_parenthesize_index(sentence)
    
    removal_idxs = list()
    
    for i, (start, last) in enumerate(idxs):
        if mode == 'odd' and i % 2 == 0:
            removal_idxs += list(range(start, last + 1))
    
    full_idxs = list(set(range(0, len(sentence))) - set(removal_idxs))
    full_idxs.sort()
    
#     print(np.asarray(sentence))
    
    filtered_sentence_as_list = np.asarray(list(sentence))[full_idxs]
    
    return ''.join(filtered_sentence_as_list)
```

```python
def remove_duplicated_space(sentence):
    
    space = ' '
    last_char = '#'
    filtered_sentence = list()
    
    for i, char in enumerate(sentence):
        if not (char == space and last_char == char):
            filtered_sentence.append(char)
            
        last_char = char
    
    return ''.join(filtered_sentence)
```

```python
def ETRI_sentence_to_jamo(sentence):
    slash_exp = re.compile('[a-z]/')
    ETRI_exp = re.compile('[*+/]')
    paren_exp = re.compile('[\(\)]')
    korean = re.compile('[^=, .?!_0-9ㄱ-ㅣ가-힣]+') 
    
    sentence = slash_exp.sub('', sentence)
    sentence = ETRI_exp.sub('', sentence)
    sentence = remove_parenthesized_word(sentence)
    sentence = paren_exp.sub('', sentence)
    sentence = remove_duplicated_space(sentence)
    sentence = sentence.strip()

    split_jamo = jamotools.split_syllables(sentence, 'JAMO')

    return split_jamo
```

```python
metadata = list()

for i, (pcm_path, txt_path) in tqdm(enumerate(zip(pcm_list, txt_list))):
    try:
        with open(pcm_path, 'rb') as pcm_file:
            pcm_data = np.fromfile(pcm_file, dtype=np.int16)
            pcm_data = pcm_data / 2 ** 15

        with open(txt_path, 'r', encoding='euc-kr') as txt_file:
            txt_data = txt_file.read()

            split_jamo = ETRI_sentence_to_jamo(txt_data)
            
        metadata.append([pcm_path, split_jamo])
    
    except:
        print("Error found in {}".format(pcm_path))
```

```python
np.save(data_path + 'ETRI_metadata.npy', metadata)
```
