```python
import numpy as np
import jamotools
from tqdm import tqdm_notebook as tqdm
import re
import glob
```

```python
data_path = '../data/aihub_corpus/'

hangul_range = range(0x1100, 0x11FF + 1)

eumjul_file_list = glob.glob(data_path + '*eumjul_list.npy')
eumun_file_list = glob.glob(data_path + '*eumun_list.npy')

eumjul_file_list.sort()
eumun_file_list.sort()
```

```python
eumjul_list = list()
eumun_list = list()

for eumjul_file, eumun_file in zip(eumjul_file_list, eumun_file_list):
    eumjul = np.load(eumjul_file)
    eumun = np.load(eumun_file)
    
    eumjul_list.append(eumjul)
    eumun_list.append(eumun)
    
eumjuls = np.concatenate(eumjul_list)
eumuns = np.concatenate(eumun_list)
```

```python
with open(scripts[9], 'r', encoding = 'utf-8') as f:
    line = f.read()
    line = line.strip()
```

```python
import jamotools
jamo = jamotools.split_syllables(line, 'JAMO')
```

```python
import re
jamo_regex = re.compile(u'[,_ ^.?!？~<>:;/%()+A-Za-z0-9\u1100-\u115e\u1161-\u11A7\u11a8-\u11ff]+')
jamo_filtered = ''.join(jamo_regex.findall(jamo))
```

```python
korean = re.compile(u'[^, .?!ㄱ-ㅣ가-힣]+') 
# korean = re.compile(u'[^, .?!\u1100-\u115e\u1161-\u11A7\u11a8-\u11ff]+') 
korean = re.compile(u'[,_ .^?!~<>:;/%()+A-Za-z0-9\u1100-\u115e\u1161-\u11A7\u11a8-\u11ff]+')
```

```python
eumjul_eumun_pairs_list = list()

for i, (eumjul, eumun) in tqdm(enumerate(zip(eumjuls, eumuns))):

    eumjul_leftovers = korean.findall(eumjul)
    eumun_leftovers = korean.findall(eumun)
    
    print(eumjul_leftovers)
    print(eumun_leftovers)
    
    
    if len(eumjul_leftovers) != 0 or len(eumun_leftovers) != 0:
        print(eumjul)
    else:
        eumjul_eumun_pairs_list.append([eumjul, eumun])
        
    break
```

```python
last_eumjul = ''
last_eumun = ''

eumjul_eumun_pairs_list_filtered = list()

for pairs in tqdm(eumjul_eumun_pairs_list):
    eumjul = pairs[0]
    eumun =  pairs[1]

    if not (eumjul == last_eumjul or eumun == last_eumun):
        if np.abs(len(eumjul) - len(eumun)) < 12:
            eumjul_eumun_pairs_list_filtered.append(pairs)
           
    last_eumjul = eumjul
    last_eumun = eumun
    
print(len(eumjul_eumun_pairs_list))

print(len(eumjul_eumun_pairs_list_filtered))
```
